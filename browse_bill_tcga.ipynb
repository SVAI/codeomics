# Explore Bill-TCGA Matches

```python
import pandas
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

CSV_PATH = "datasets/results/bill_tcga.csv"
```

```python
df = pandas.read_csv(CSV_PATH)
print("Count: ", len(df))
df
```

## TCGA Variant Classification

```python
print("Unique Values: ", len(df['tcga_Variant_Classification'].unique()))
```

```python
df[['tcga_Variant_Classification', 'Count']]\
    .groupby(by='tcga_Variant_Classification').sum()\
    .nlargest(10, 'Count')\
    .plot.pie(figsize=(12, 12), subplots=True);
        
```

## TCGA Variant Type

```python
print("Unique Values: ", len(df['tcga_Variant_Type'].unique()))
```

```python
df[['tcga_Variant_Type', 'Count']]\
    .groupby(by='tcga_Variant_Type').sum()\
    .nlargest(10, 'Count')\
    .plot.pie(figsize=(12, 12), subplots=True);
        
```

## TCGA Project Short Name

```python
df[['tcga_project_short_name', 'Count']].groupby(by='tcga_project_short_name').sum()\
    .nlargest(10, 'Count')\
    .plot.bar(figsize=(14, 8));
```

## TCGA Hugo Symbol

```python
print("Unique Values: ", len(df['tcga_Hugo_Symbol'].unique()))
```

```python
df[['tcga_Hugo_Symbol', 'Count']].groupby(by='tcga_Hugo_Symbol').sum()\
    .nlargest(10, 'Count')\
    .plot.bar(figsize=(14, 8));
```

## TCGA Hugo Symbol & Project Short Name

```python
top_genes = df[df['tcga_Hugo_Symbol'] != 'Unknown'][['tcga_Hugo_Symbol', 'Count']].groupby(by='tcga_Hugo_Symbol').sum()\
    .nlargest(10, 'Count')
top_genes.index.values
```

```python
grid = df[df['tcga_Hugo_Symbol'].isin(top_genes.index.values)][['tcga_Hugo_Symbol', 'tcga_project_short_name', 'Count']]\
    .groupby(by=['tcga_Hugo_Symbol', 'tcga_project_short_name']).sum()\
    .unstack().fillna(0)
# grid  
```

```python
grid.columns = grid.columns.get_level_values(1)
grid.plot.bar(stacked=True, figsize=(16, 12));
```
