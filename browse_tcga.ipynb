# Explore TCGA Stats

```python
import pandas
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

## TCGA Variant Classification

```python
# Data from BigQuery
df = pandas.DataFrame.from_dict([
  {
    "Count": "2084733",
    "Variant_Classification": "Missense_Mutation"
  },
  {
    "Count": "1090899",
    "Variant_Classification": "Frame_Shift_Del"
  },
  {
    "Count": "875801",
    "Variant_Classification": "Silent"
  },
  {
    "Count": "275523",
    "Variant_Classification": "RNA"
  },
  {
    "Count": "224678",
    "Variant_Classification": "Frame_Shift_Ins"
  },
  {
    "Count": "164325",
    "Variant_Classification": "Intron"
  },
  {
    "Count": "153628",
    "Variant_Classification": "Nonsense_Mutation"
  },
  {
    "Count": "151076",
    "Variant_Classification": "Splice_Site"
  },
  {
    "Count": "139504",
    "Variant_Classification": "lincRNA"
  },
  {
    "Count": "42822",
    "Variant_Classification": "3'UTR"
  }
])
df["Count"] = df["Count"].astype(int)
df = df.set_index('Variant_Classification')
```

```python
df.groupby(by='Variant_Classification').sum()\
    .nlargest(10, 'Count')\
    .plot.pie(figsize=(12, 12), subplots=True);
        
```
